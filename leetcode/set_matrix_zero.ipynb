```python
'''
iterate through the array and store the indices of rows and columns that satisfies the criteria and has to be converted to zero
iterate through the   elements at stored indices and update them to zer

'''
class Solutions:
    def  setZero(self, matrix):
        rows_to_zero = []
        cols_to_zero = []

        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                if  matrix[i][j] == 0:
                    rows_to_zero.append(i)
                    cols_to_zero.append(j)
        for i in range(len(matrix)):
            for j in range(len(matrix[0])):
                for i in rows_to_zero or cols_to_zero:
                    matrix[i][j] = 0
        return matrix

self = Solutions()
matrix = [[1,1,1],[1,0,1],[1,1,1]]
zeroed_MATRIX = self.setZero(matrix)
print(zeroed_MATRIX)

```
