```python
'''
prefix sum also called as cumulative sum is an array where ith elements holds the summ of all the previous elements starting from 0th index to ith index
for example
nums = [a, b, c, d]
prefix[i] = a + b + c + ... + nums[i]

why prefix sum is used:
while computing the sum over a range , linear summation is  computationally expensive
'''
nums = [1,2,3,4]
nums
```
