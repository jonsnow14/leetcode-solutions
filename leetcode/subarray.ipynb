Formula for number of subarrays

For an array of length n, the total number of contiguous subarrays is:

Total subarrays = n × (n + 1) / 2

#solution approach
prefixSum[j] − prefixSum[i] is divisible by k

if prefix sum upto ith index and prefix sum upto jth index (j<i) are pi and pj ,  then subarray is  divisible by k if (pi-pj)%k = 0


maintain  a running current_sum
calculate the remainder of current_sum divided by k
check the remainder_map, checking how many times specific remainder has occured previously
if seen X times previously add X to  total count(becuase the current index can form a valid subarray with all X previous indices
update the map to include the current remainder

```python
class Solution:

    def subarraysDivByK(self, nums, k):
        current_sum = 0
        n = len(nums)
        count = 0
        remainder_map = {0:1}

        for num in nums:
            current_sum += num
            remainder = current_sum %k
            if remainder in remainder_map:
                count+=remainder_map[remainder]
                remainder_map[remainder]+=1
            else:
                remainder_map[remainder]=1

        return count
self = Solution()
nums = [1,2,3,4,5,6,7]
result = self.subarraysDivByK(nums,3)
print(result)


```
